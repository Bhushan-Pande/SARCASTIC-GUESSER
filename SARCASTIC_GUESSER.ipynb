```python
#TRAINING
# ON NEWS HEADLINE

import json

with open('/content/drive/MyDrive/sar_xy.json','r') as f:
  datastore=json.load(f)
# print(datastore)
sentences=[]
labels=[]
urls=[]

for item in datastore:
    sentences.append(item['headline'])
    labels.append(item['is_sarcastic'])
    urls.append(item['article_link'])
  
```

```python
vocab_size = 100000
embedding_dim = 16
max_length = 100
trunc_type='post'
padding_type='post'
oov_tok = "<OOV>"
training_size = 20000
training_sentences=sentences[:20000]
test_sentences=sentences[20000:]
training_labels=labels[:20000]
test_labels=labels[20000:]

```

```python
tokenizer=Tokenizer(num_words=100000,oov_token="<OOV>")
tokenizer.fit_on_texts(training_sentences)
word_index=tokenizer.word_index
```

```python
training_sentences_sequences=tokenizer.texts_to_sequences(training_sentences)
training_sentences_padded=pad_sequences(training_sentences_sequences,padding='post',truncating=trunc_type)
# training_labels_sequences=tokenizer.texts_to_sequences(training_labels)
# training_labels_padded=pad_sequences(training_labels_sequences,padding='post')
```

```python
test_sentences_sequences=tokenizer.texts_to_sequences(test_sentences)
test_sentences_padded=pad_sequences(test_sentences_sequences,padding='post',truncating=trunc_type)
# test_labels_sequences=tokenizer.texts_to_sequences(test_labels)
# test_labels_padded=pad_sequences(test_sequences_labels_sequences,padding='post')
```

```python
import numpy as np
training_padded = np.array(training_sentences_padded)
training_labels = np.array(training_labels)
testing_padded = np.array(test_sentences_padded)
testing_labels = np.array(test_labels)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, input_length=max_length),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(24, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
num_epochs = 30
history = model.fit(training_padded, training_labels, epochs=num_epochs, validation_data=(testing_padded, testing_labels), verbose=2)
```

```python
sentence = ["granny starting to fear spiders in the garden might be real", "I’m not saying I hate you, what I’m saying is that you are literally the Monday of my life","Find your patience before I lose mine."]
sequences = tokenizer.texts_to_sequences(sentence)
padded = pad_sequences(sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)
print(model.predict(padded))
```
